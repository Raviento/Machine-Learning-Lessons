```python
import warnings
warnings.filterwarnings('ignore')

from prophet import Prophet
import yfinance as yf
```

```python
df=yf.download('GOOG',"2019-01-01","2025-04-26")
```

```python
df=df[['Close']]
```

```python
df=df.reset_index()
```

```python
df.columns=['ds','y']
```

```python
df.head(3)
```

```python
model=Prophet()
```

```python
model.fit(df)
```

```python
gelecek=model.make_future_dataframe(360)
```

```python
tahmin=model.predict(gelecek)
```

```python
model.plot(tahmin);
```

```python
model.plot_components(tahmin);
```
