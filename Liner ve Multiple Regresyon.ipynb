#  Linear Regression & Multiple Linear Regression (Dorusal ve oklu Dorusal Regresyon)

---



##  Nedir?

**Linear Regression (Dorusal Regresyon)**, bir deikenin ( hedef  *dependent variable*) baka bir deikenle ( zellik  *independent variable*) olan **dorusal ilikiyi** modellemeye yarar.

 rnein:  
Bir evin bykl (m) arttka, fiyat nasl deiir?  

---

##  Temel Formller

---

###  Basit (Tek Deikenli) Dorusal Regresyon

$$
y = \beta_0 + \beta_1 x + \varepsilon
$$

- \( y \): Tahmin edilen sonu (r: fiyat)  
- \( x \): Girdi deikeni (r: byklk)  
- \( \beta_0 \): Sabit terim (*intercept*)  
- \( \beta_1 \): Eim / katsay (*slope*)  
- \( \varepsilon \): Hata pay (*noise*)

---

###  Multiple Linear Regression (oklu Dorusal Regresyon)

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_p x_p + \varepsilon
$$

Birden fazla deiken varsa (r. byklk, oda says, bina ya), hepsi birlikte \( y \)yi etkiler.  
**Model, tm bu zellikleri dikkate alarak en uygun tahmini yapar.**


---
