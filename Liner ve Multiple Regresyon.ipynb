# 📊 Linear Regression & Multiple Linear Regression (Doğrusal ve Çoklu Doğrusal Regresyon)

---



## 🔹 Nedir?

**Linear Regression (Doğrusal Regresyon)**, bir değişkenin (🎯 hedef – *dependent variable*) başka bir değişkenle (📏 özellik – *independent variable*) olan **doğrusal ilişkiyi** modellemeye yarar.

👉 Örneğin:  
“Bir evin büyüklüğü (m²) arttıkça, fiyatı nasıl değişir?”  

---

## 🔹 Temel Formüller

---

### 🧮 Basit (Tek Değişkenli) Doğrusal Regresyon

$$
y = \beta_0 + \beta_1 x + \varepsilon
$$

- \( y \): Tahmin edilen sonuç (ör: fiyat)  
- \( x \): Girdi değişkeni (ör: büyüklük)  
- \( \beta_0 \): Sabit terim (*intercept*)  
- \( \beta_1 \): Eğim / katsayı (*slope*)  
- \( \varepsilon \): Hata payı (*noise*)

---

### 🧮 Multiple Linear Regression (Çoklu Doğrusal Regresyon)

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_p x_p + \varepsilon
$$

Birden fazla değişken varsa (ör. büyüklük, oda sayısı, bina yaşı), hepsi birlikte \( y \)’yi etkiler.  
**Model, tüm bu özellikleri dikkate alarak en uygun tahmini yapar.**


---

## 🔹 Temel Kavramlar

| 📌 İngilizce Terim | 🇹🇷 Türkçesi | 💡 Anlamı |
|--------------------|-------------|-----------|
| Dependent Variable | Bağımlı değişken | Tahmin edilen sonuç |
| Independent Variable | Bağımsız değişken | Tahmin için kullanılan özellik |
| Intercept (β₀) | Sabit terim | Modelin başlangıç noktası |
| Coefficient (β₁, β₂…) | Katsayı | Özelliğin etkisi (1 birim artarsa y ne kadar değişir?) |
| Residual | Artık (hata) | Gerçek - tahmin farkı |
| MSE | Ortalama Kare Hatası | Hataların karelerinin ortalaması |
| R² | Açıklanan varyans oranı | Model veriyi ne kadar iyi açıklıyor? (0–1 arası) |

---

## 🔹 Ne Zaman Kullanılır?

✅ Sürekli (numerik) bir şeyi tahmin ediyorsan (ör: fiyat, maaş, sıcaklık)  
✅ Özelliklerin ve hedefin **doğrusal ilişki** içindeyse  
✅ Basit, hızlı ve yorumlanabilir bir model istiyorsan  

---

## 🔍 Mini İpucu: 
> Çok sayıda değişkenle çalışıyorsan **Multiple Linear Regression**,  
> Tek değişkenle çalışıyorsan **Simple Linear Regression** uygundur.
