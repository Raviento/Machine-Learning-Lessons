```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('twd.csv')
df.head(3)
```

```python
df['Team'].unique() # Takım isimlerini sıralar
```

```python
le=LabelEncoder()
le.fit(df['Team'])
```

```python
x=le.transform(df['Team']) # Kategorisel verileri sayısal verilere çevirdik
```

```python
le.classes_
```

```python
x
```

```python
le.inverse_transform(x) # Eğer sayısal verileri yine kategorisel verilere çevirmek istersek bunu kullanıyoruz.
```

```python
le2=LabelEncoder()
le2.fit_transform(df['Team']) # Kısayoldan yapmak için bu kod parçasını kullanabiliriz.
```
