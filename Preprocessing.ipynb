```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Audi_A1_listings.csv")
```

```python
df.head(3)
```

```python
df = df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score']) # Lazım olmayan sütunları siliyoruz preprocessing
```

```python
df.head(2) # Görüldüğü üzere hem Numeric kısımlar (ör. year) hemde Kategorisel kısımlar (ör. fuel) var bunları ayırmak lazım
```

```python
df.info() # floatlar numeric objectler kategorisel
```

```python
df.columns=['yil', 'kasa', 'mil', 'motor', 'ps', 'vites', 'yakit', 'sahip', 'fiyat', 'ppy'] # sütunların isimlerini anlaşılır yapıyoruz
```

```python
df.head(2)
```

```python
df['motor'] = df['motor'].str.replace("L","") # sonundaki L harfini silmek için
```

```python
df['motor'].info() # göründüğü üzere hala object numeric yapmamız lazım
```

```python
pd.to_numeric(df['motor']) # motor kolonunu numeric yapıyoruz
```

```python
pd.get_dummies(df,columns=['kasa','vites','yakit']) #Kategorisel verileri numeric tipe çevirir
```

```python
df =pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True) # İse sütun sayılarını azaltır ve basitleştirir mesela dizelse 1 petrolse 0
```

```python
x = df.drop('fiyat',axis=1) # fiyat sütunu hariç diğer sütunları dahil ediyor
y = df[["fiyat"]] # Fiyatı tahmin edeceğiz
```

```python
lm = LinearRegression()
model = lm.fit(x,y) # Modeli eğitiyoruz
```

```python
model.predict([[2017,30000,1.6,110,1,2600,0,1]]) # Girdilere göre tahmin ediyoruz
```

```python
model.score(x,y) # Başarı skoru
```
